```python
import sys
sys.path.append('./../..')
import dynaphos
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from dynaphos import utils
import seaborn as sns
import matplotlib as mpl


results = pd.read_csv('results/performance_best_GPU.csv',)
# Load colorstyles 
plt.style.use('../_figurestyle/small.mplstyle')
COLORSTYLE = utils.load_params('../_figurestyle/colors.yaml')

CMAP = 'Reds' # 'Reds'#'OrRd'#'gist_heat_r'
```

```python



f, ax = plt.subplots(figsize=(4, 2),dpi=300)

# sns.barplot(data= results, x='n_phosphenes', y='fps', hue='res-tag', ax=ax)
sns.barplot(data=results, hue='n_phosphenes', y='fps', x='res-tag', ax=ax,
            palette=sns.color_palette(CMAP, 7,desat=0.9),
            errwidth = 0.4,
            capsize = 0.07
           
           )
ax.set(xlabel = 'Resolution',
       ylabel = 'Frames Per Second',
#        ylim = [None,50000000000],
       xlim = [None,5.2],
       yscale="log", 
       xticklabels = [r'${} \times{} $'.format(n,n) for n in [64,128,256,512,1024] ]
        )

ax.legend(title='No. Phosphenes', ncol=1,
#           loc='lower left',
           loc='upper right',
          bbox_to_anchor=(1.05, 1),
#           alpha = 0.1
          fancybox=True,
          framealpha=0.0
          
         )

empty = [3.25,  4.125, 4.25 ]  #[2.25, 3.125, 3.25, 4., 4.125, 4.25 ] 
ax.plot(empty, np.ones_like(empty)*2, 'kx', markersize=3)
```
